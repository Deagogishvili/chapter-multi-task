# Import Libraries

```python
import warnings
warnings.filterwarnings("ignore")
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joypy
from joypy import joyplot
from matplotlib import cm
```

# Define paths

```python
data_path = os.path.dirname(os.getcwd()) + '/data'
figure_path = os.path.dirname(os.getcwd()) + '/figures'
```

```python
PROT = pd.read_csv(data_path + '/predicted/human_proteome.csv')
AF = pd.read_csv(data_path + '/predicted/alphafold_molpatch.csv', sep=',', engine='python') # Molpatch output on AlphaFold
old_molpatch = pd.read_csv(data_path + '/predicted/molpatch_old.csv', sep=',', engine='python') # Old MolPatch
```

```python
consensus = pd.read_csv(data_path + '/predicted/rna_tissue_consensus.tsv', sep='\t', engine='python') 
TPM = consensus.sort_values('nTPM').drop_duplicates(subset=['Gene'], keep='last') 
# log tansform expression values
TPM['TPM_LOG2'] = np.log2(TPM['nTPM']+1)

mapped = pd.read_csv(data_path + '/predicted/Consensus_mapped.txt', sep='\t', engine='python') 
mapped['id'] = mapped['To']
mapped['Gene name'] = mapped['From']
mapped = mapped[['Gene name', 'id']]
TPM = TPM.merge(mapped, on='Gene name')
TPM['ID'] = TPM['id']
PROT = PROT.merge(TPM, on='ID')
```

```python
# log tansform expression values
PROT['LHP_LOG2'] = np.log2(PROT['LHP_predicted']+1)
```

```python
PROT
```

```python
#sns.set_style(style='white')
fig , ax = plt.subplots(figsize = (17,8))
sns.set(font_scale = 1.55)

cm = plt.cm.get_cmap('viridis')
PROT = PROT.sort_values(by=['TPM_LOG2'], ascending=True)

x = PROT['LHP_LOG2']
y = PROT['RHSA_predicted']
colors = PROT['TPM_LOG2'] 

im = plt.scatter(x, y, s=60, c=colors, alpha=0.7) #, cmap=cm)
fig.colorbar(im, ax=ax, label='Normalised Expression (LOG2)')
ax.set_xlabel('LHP (LOG2)', fontsize=20) #($Å^2$)
ax.set_ylabel('RHSA', fontsize=20)
ax.set_title('', fontsize=30)
plt.show()

fig.savefig(figure_path + '/LHP_expression.png', dpi = 300, bbox_inches='tight')
```

```python
AF_patches = list(AF['LHP'])
PROT_patches = list(PROT['LHP_predicted'])
MP_patches = list(old_molpatch['size'])

# Create a list of method names corresponding to the data points
methods = ["PatchProt"] * len(PROT_patches) + ["AlphaFold"] * len(AF_patches) + ["MolPatch"] * len(MP_patches)

# Combine the data and method names into a single DataFrame
joyplot_df = pd.DataFrame({'Method': methods, 'LHP': PROT_patches+ AF_patches + MP_patches})

# log tansform values
joyplot_df['LHP_LOG2'] = np.log2(joyplot_df['LHP']+1)
```

```python
# Create the joyplot
sns.set(font_scale = 1.55)
fig, axes = joypy.joyplot(joyplot_df, by="Method", column="LHP_LOG2", colormap=cm.Pastel1, figsize=(10, 5), alpha=0.8)

# Set the figure background color to white
fig.patch.set_facecolor('white')

# Set the axis background color to white for each axis in the list
for ax in axes:
    ax.set_facecolor('white')

# Add an x-label
axes[-1].set_xlabel('Large Hydrophobic Patch (LOG2)')
fig.savefig(figure_path + '/patches_01.png', dpi = 300, bbox_inches='tight')
plt.show()
```

```python
fig, axes = joypy.joyplot(joyplot_df, by="Method", column="LHP", colormap=cm.Pastel1, figsize=(10, 5), alpha = 0.8)
fig.patch.set_facecolor('white')
for ax in axes:
    ax.set_facecolor('white')
axes[-1].set_xlabel('Large Hydrophobic Patch ($Å^2$)')
fig.savefig(figure_path + '/patches_02.png', dpi = 300, bbox_inches='tight')
plt.show()
```

```python
sns.set_style('ticks')
sns.set(font_scale = 1.55)
fig, ax = plt.subplots(1,1)
fig.set_size_inches(17,8)

plt.hist(AF['LHP'], color='lightgreen', ec='black', bins=35, alpha=0.5)
plt.hist(PROT['LHP_predicted'], color='crimson', ec='black', bins=25, alpha=0.5)
plt.hist(old_molpatch['size'], color='blue', ec='black', bins=5, alpha=0.5)
legend = ['MolPatch AlphaFold','LHP Predictions', 'MolPatch X-ray PDB']
plt.legend(legend, prop={'size': 20}) #, title = 'Data Set')
plt.title('',  size=30)
plt.xlabel('LHP ($Å^2$)',  size=25)
plt.ylabel('Density',  size=25)
```
